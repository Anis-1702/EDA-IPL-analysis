```python
plt.figure(figsize= (12,8))
df['winner'].value_counts().plot(kind = 'bar' ,colormap = 'summer' )
plt.title('Number of winnings on Team wise')
plt.ylabel("no of matches")
plt.show()
```

INTERPREATATION :

Highest number of matches won by MUMBAI INDIAN

LOWEST number of matches won by RISING PUNE SUPERGIANTS

```python
plt.figure(figsize= (12,8))
df['player_of_match'].value_counts()[:10].plot(kind = 'bar' , colormap = "summer")
plt.title('TOP 10 PLAYER MAN OF THE MATCH')
plt.ylabel('number of matches')
plt.xticks(rotation = 100)
plt.show()
```

INTERPREATATION :

Highest number of man of the match award won by CH Gayle

```python
toss_win_match_win= df[df['toss_winner'] == df['winner']]
```

```python
print(f"{len(toss_win_match_win)} times a team won the toss won the match and {len(df)-len(toss_win_match_win)} times a team won the toss did not win the match")
```

```python
plt.figure(figsize= (12,8))
df[df['winner']== 'Chennai Super Kings'].venue.value_counts().plot(kind = 'bar' )
plt.title ('CSK wining matches based venue')
plt.ylabel('number of matches')
plt.show()
```
